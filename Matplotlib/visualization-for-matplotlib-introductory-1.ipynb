# **<span style="color:#4686C8;">Types of inputs to plotting functions</span>**

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10001)
data = {'a' : np.arange(50),
        'c' : np.random.randint(0, 50, 50),
        'd' : np.random.randn(50)}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d'])*100

fig, ax = plt.subplots(figsize = (5, 2.5), layout = 'constrained')
# You can using Dictionary about data
ax.scatter('a', 'b', c = 'c', s = 'd', data = data)
ax.set_xlabel('Entry A')
ax.set_ylabel('Entry B')
plt.show()
```

# **<span style="color:#4686C8;">Helping function for Visualization</span>**

**If you need to make same plot and diffenrent dataset, or want to easily wrap Matplotlib methods,**

**we use the recommended signature function below.**

```python
def helping_function(ax, data1, data2, param_dict):
    """
    It will help Making same plot and diffent dataset.
    """
    out = ax.plot(data1, data2, **param_dict)  
    return out  # return about Plot

data1, data2, data3, data4 = np.random.randn(4, 100)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (8, 5))   # row, cols
helping_function(ax1, data1, data2, {'marker' : 'x'})
helping_function(ax2, data3, data4, {'marker' : 'o'})
plt.show()
```

# **<span style="color:#4686C8;">Axes labels and text</span>**

**set_xlabel, set_ylabel and set_title are used to add text in the indicated locations**

**If you want using mathematical expressions in text, Matplotlib accepts TexX equation expressions in any text expression.**

**You can reference two LINK**

<a href = "https://matplotlib.org/stable/tutorials/text/mathtext.html">LINK 1</a>

<a href = "https://matplotlib.org/stable/tutorials/text/usetex.html">LINK 2</a>

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize = (8, 5), layout = 'constrained')

n, bins, patches = ax.hist(x, 50, density = 1, facecolor = 'C0', alpha = 0.75)
ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title('Kaggle Tutorial\n (not really)')
ax.text(75, .025, r'$\mu=115,\ \sigma=15$', fontsize = 15.0)
ax.axis([55, 175, 0, 0.03])
ax.grid(True);
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (7, 4))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, lw = 2)

ax.annotate('local max', xy = (2, 1), xytext = (3, 1.5),
            arrowprops = dict(facecolor = 'black', shrink = 0.05))

ax.set_ylim(-2, 2)
plt.show()
```

# **<span style="color:#4686C8;">Additional Axis objects</span>**

**Plotting data of different magnitude in one chart may require an additional y-axis**

**Using twinix add a new Axes with an invisible x-axis and y-axis positioned at the right**

**Similarly, you can add a secondary_xaxis or secondary_yaxis having a different scale than the main axis to reperesnt the data in different scales or units.**

```python
fig, (ax1, ax3) = plt.subplots(1, 2, figsize = (7, 2.7), layout = 'constrained')
l1, = ax1.plot(t, s)
ax2 = ax1.twinx()
l2, = ax2.plot(t, range(len(t)), 'C1')
ax2.legend([l1, l2], ['Sine (left)', 'Straight (right)'])

ax3.plot(t, s)
ax3.set_xlabel('Angle [rad]')
ax4 = ax3.secondary_xaxis('top', functions = (np.rad2deg, np.deg2rad))
ax4.set_xlabel('Angle [*]')
```

# **<span style="color:#4686C8;">Working with multiple Figures and Axes</span>**

**You can open multiple Figures with multiple calls to fig = plt.figure() or fig2, ax = plt.subplots().**

**Multiple Axes can be added a number of ways, but the most basic is plt.subplots() as used above.**

```python
fig, axd = plt.subplot_mosaic([['1', '3'],
                               ['2', '3']], layout='constrained')
axd['1'].set_title('1')
axd['2'].set_title('2')
axd['3'].set_title('3');
```
