# **<span style="color:#4686C8;">Intro to Pyplot</span>**

**Two Method**
- Categorical Variables [stateless API] :  It is the method of drawing on the figure I designated, the ax I designated.
- Categorical Variables [stateful API] : This method automatically finds the current figure and ax and plots it there.

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# Stateless API 
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

fig, ax = plt.subplots(1, 3, figsize = (9, 3))

ax[0].bar(names, values)
ax[0].grid(True)
ax[1].scatter(names, values)
ax[2].plot(names, values)
ax[2].grid(True)
fig.suptitle("Category Plotting", fontsize = 15)
plt.show()
```

```python
# Stateful API

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.grid(True)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.grid(True)
plt.plot(names, values)
plt.suptitle('Categorical Plotting', fontsize = 15)
plt.show()
```

# **<span style="color:#4686C8;">Logarithmic and other nonlinear axes</span>**

**matplotlib.pyplot supprots not only linear axis scales, but also logarithmic and logit scales.**

**This is commonly used if data spans many orders of magnitude.Changing the scale of an axis is easy.**

<a href= "https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots_adjust.html#matplotlib.pyplot.subplots_adjust">Subplots_adjust Parameters</a> : It's useful Making Multiple Matplotlib

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.35, bottom=-1, left=0.10, right=0.95, hspace=0.55,
                    wspace=0.35)

plt.show()
```

# **<span style="color:#4686C8;">Combining multiple visualizations</span>**

**It's possible to draw multiple plot elements on the same instance of axes.Axes. To do this we simply need to call another one of the plot methods on the axes object.**

**Don't forget It's different plt.setp and ax.set!!!**

- set_major_formatter : Set the formatter of the major ticker. Parameters : formatter : Formatter, str, or function
<a href= "https://matplotlib.org/stable/api/_as_gen/matplotlib.axis.Axis.set_major_formatter.html#matplotlib.axis.Axis.set_major_formatter">LINK</a>

```python
def currency(x, pos):
    """The two arguments are the value and tick position"""
    if x >= 1e6:
        s = '${:1.1f}M'.format(x*1e-6)
    else:
        s = '${:1.0f}K'.format(x*1e-3)
    return s
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation = 45, horizontalalignment = 'right')

ax.axvline(group_mean, ls = '--', color = 'r')

for group in [3, 5, 8]:
    ax.text(145000, group, "New Company", fontsize =10, verticalalignment = 'center')

ax.title.set(y = 1.05)
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
ax.xaxis.set_major_formatter(currency)
ax.set_xticks([0, 25e3, 50e3, 75e3, 100e3, 125e3])

plt.show()
```

# **<span style="color:#4686C8;">Customizing Matplotlib with style sheets and rcParams</span>**

**There are three ways to customize Matplotlib.**
- Setting rcParams at runtime.
- Using style sheets.
- Changing your matplotlibrc file.  <a href = "https://matplotlib.org/stable/tutorials/introductory/customizing.html#matplotlibrc-sample">LINK</a>

Setting rcParams at runtim takses precedence over style sheets take pecdence over matplotlibrc files.

**You can define your stylesheet in matplotlib. check! <a href ="https://matplotlib.org/stable/tutorials/introductory/customizing.html#matplotlibrc-sample">LINK</a>**

```python
fig, ax = plt.subplots(figsize = (10, 6))
mpl.rcParams['lines.linewidth'] = 5
mpl.rcParams['lines.linestyle'] = '--'
data = np.random.randn(50)
plt.plot(data, color = 'blue')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
mpl.rcParams['lines.linewidth'] = 5
mpl.rcParams['lines.linestyle'] = '--'
data = np.random.randn(50)
plt.plot(data, color = 'red')
plt.show()
```

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```
